```python
class Deck(object):
    
    def __init__(self, dck):
        self.dck=dck
        
    def shuffle_deck(self):
        shuffle(self.dck)
        return self.dck
    
    def deal_player(self):
        return self.dck[0:2]
    
    def deal_dealer(self):
        return self.dck[2:4]
```

```python
class Banking(object):
    
    def __init__(self):
        bankroll = 100
        
    def deposit(self):
        while True:
            bankroll = int(raw_input("How much money are you starting out with? ($10 min, $100000 max): $"))
            if bankroll < 10.00:
                print "That's too little"
            elif bankroll > 100000.00:
                print "That's too much."
            elif 10.00<=bankroll<=100000.00:
                break
            else:
                print "Try again"
        return bankroll
    
    def betting(self,bankroll):
        while True:
            bet = int(raw_input("How much is your bet? $"))
            if bet < 1.00:
                print "That's too little."
            elif bet > bankroll:
                print "You don't have enough money for that."
            elif 1<=bet<=bankroll:
                break
            else:
                print "Try again"
        return bet
                
    def settle_bet(self,bankroll,bet,win_or_lose):
        if win_or_lose == True:
            bankroll += bet
        else:
            bankroll -= bet
        return bankroll
    
    def check_broke(self,bankroll):
        if bankroll == 0:
            print "You are out of money."
            return False
        else:
            return True
```

```python
def create_deck():
        deck=[]
        for i in range(4):
            for x in range(2,15):
                if i==0:
                    if x==11:
                        deck.append("Jack of Hearts")
                    elif x==12:
                        deck.append("Queen of Hearts")
                    elif x==13:
                        deck.append("King of Hearts")
                    elif x==14:
                        deck.append("Ace of Hearts")
                    else:
                        deck.append(str(x)+ ' of Hearts')
                elif i==1:
                    if x==11:
                        deck.append("Jack of Hearts")
                    elif x==12:
                        deck.append("Queen of Hearts")
                    elif x==13:
                        deck.append("King of Hearts")
                    elif x==14:
                        deck.append("Ace of Hearts")
                    else:
                        deck.append(str(x)+' of Diamonds')
                elif i==2:
                    if x==11:
                        deck.append("Jack of Hearts")
                    elif x==12:
                        deck.append("Queen of Hearts")
                    elif x==13:
                        deck.append("King of Hearts")
                    elif x==14:
                        deck.append("Ace of Hearts")
                    else:
                        deck.append(str(x)+' of Spades')
                else:
                    if x==11:
                        deck.append("Jack of Hearts")
                    elif x==12:
                        deck.append("Queen of Hearts")
                    elif x==13:
                        deck.append("King of Hearts")
                    elif x==14:
                        deck.append("Ace of Hearts")
                    else:
                        deck.append(str(x)+' of Clubs')
        return deck                       
```

```python
def hand_value(lst):
    values=[]
    for x in lst:
        if x == '2 of Hearts' or x=='2 of Diamonds'or x=='2 of Spades'or x=='2 of Clubs':
            values.append(2)
        elif x == '3 of Hearts' or x=='3 of Diamonds'or x=='3 of Spades'or x=='3 of Clubs':
            values.append(3)
        elif x == '4 of Hearts' or x=='4 of Diamonds'or x=='4 of Spades'or x=='4 of Clubs':
            values.append(4)
        elif x == '5 of Hearts' or x=='5 of Diamonds'or x=='5 of Spades'or x=='5 of Clubs':
            values.append(5)
        elif x == '6 of Hearts' or x=='6 of Diamonds'or x=='6 of Spades'or x=='6 of Clubs':
            values.append(6)
        elif x == '7 of Hearts' or x=='7 of Diamonds'or x=='7 of Spades'or x=='7 of Clubs':
            values.append(7)
        elif x == '8 of Hearts' or x=='8 of Diamonds'or x=='8 of Spades'or x=='8 of Clubs':
            values.append(8)
        elif x == '9 of Hearts' or x=='9 of Diamonds'or x=='9 of Spades'or x=='9 of Clubs':
            values.append(9)
        elif x == '10 of Hearts' or x=='10 of Diamonds'or x=='10 of Spades'or x=='10 of Clubs' or\
            x == 'Jack of Hearts' or x=='Jack of Diamonds'or x=='Jack of Spades'or x=='Jack of Clubs' or\
            x == 'Queen of Hearts' or x=='Queen of Diamonds'or x=='Queen of Spades'or x=='Queen of Clubs' or\
            x == 'King of Hearts' or x=='King of Diamonds' or x=='King of Spades'or x=='King of Clubs':
                values.append(10)
        elif x == 'Ace of Hearts' or x=='Ace of Diamonds'or x=='Ace of Spades'or x=='Ace of Clubs':
            values.append(11)
    
    value = [0]
    for x in values:
        p=0
        z=0
        if x == 11:
            for i in value:
                value[p]=value[p]+x
                p+=1
            value.append(value[len(value)-1]-10)
        else:
            for i in value:
                value[z]=value[z]+x
                z+=1
            
    return value   
```
